```python
import gradio as gr

def form_examples(first_name, last_name, email):
  return f"Form submitted successfully! Welcome, {first_name}{last_name}. Your email is {email}."

with gr.Blocks() as demo:
  with gr.Group():
    gr.Markdown("## User Information")
    with gr.Row():
      first_name = gr.Textbox(label="First Name", placeholder = "Enter your first name")
      last_name = gr.Textbox(label="Last Name", placeholder = "Enter your last name ")
    email = gr.Textbox(label="Email", placeholder = "Enter your email")
    submit = gr.Button("Submit", variant="primary")
    submit.click(form_examples, inputs=[first_name, last_name, email], outputs=gr.Textbox())

demo.launch()
```

```python
import gradio as gr
from transformers import pipeline

translation_pipelines = {
    "English": None,
    "Spanish": pipeline("translation_en_to_es", model="Helsinki-NLP/opus-mt-en-es"),
    "French": pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr"),
    "German": pipeline("translation_en_to_de", model="Helsinki-NLP/opus-mt-en-de"),
}

def create_post(content, tags, image):
  text = f"{content}\n\n{' '.join(tags)}" # join()은 리스트를 공백으로 연결한 문자열로 만들어 줌
  return image, text

  """
  words = ["apple", "banana", "cherry"]
  print(', '.join(words))   # apple, banana, cherry
  print(' - '.join(words))  # apple - banana - cherry
  print('\n'.join(words))   # apple\nbanana\ncherry (줄바꿈)
  """

def translate(content, target_lang):
    if target_lang == "English" or not content.strip():
        return content
    translator = translation_pipelines.get(target_lang)
    if translator:
        result = translator(content, max_length=500)
        return result[0]['translation_text']
    else:
        return f"Translation not available for this {target_lang}."

with gr.Blocks() as demo:
  with gr.Group():
    gr.Markdown("### Create Post")
    image_input = gr.Image(label="Upload a Photo", type="filepath") # value 안에 원하는 이미지를 넣으면 됨
    content = gr.Textbox(label="Content", placeholder="Write your post here..", lines=3)
    tags = gr.CheckboxGroup(label="Tags", choices=['#Social', "#Life", "#Travel", "#Tech"])
    create = gr.Button("Create Post", variant = "primary")

    post_image_output = gr.Image(label="Posted Image")
    post_text_output = gr.Textbox(label="Post Text")

    create.click(create_post, inputs=[content, tags, image_input], outputs=[post_image_output, post_text_output])

    target_lang = gr.Dropdown(label="Traget Language", choices=["English", "Spanish", "French", "German"])
    translate_btn = gr.Button("Translate", variant="primary")
    output= gr.Textbox(label="Translation", interactive=False) #사용자가 직접 해당 입력창에 텍스트를 입력할 수 없도록 만듬
    translate_btn.click(translate, inputs=[content, target_lang], outputs=output)

demo.launch()
```
